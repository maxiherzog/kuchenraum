```python
import venn
import numpy as np
# binarize the features and put ids into a lists
lists = {"chocolate": [], "fruity": [], "creamy": [], "cake_type": [], "streusel": []}

for cake in cakes.itertuples():
  if cake.chocolate >= np.mean(cakes['chocolate']):
    lists['chocolate'].append(cake.Index)
  if cake.fruity >= np.mean(cakes['fruity']):
    lists['fruity'].append(cake.Index)
  if cake.creamy >= np.mean(cakes['creamy']):
    lists['creamy'].append(cake.Index)
  if cake.cake_type >= np.mean(cakes['cake_type']):
    lists['cake_type'].append(cake.Index)
  if cake.streusel >= np.mean(cakes['streusel']):
    lists['streusel'].append(cake.Index)

print([*lists.values()])

labels = venn.get_labels(list(lists.values()))
fig, ax = venn.venn5(labels, names=list(lists.keys()))
fig.show()
```

```python
import networkx as nx

G = nx.Graph()

for cake in cakes.itertuples():
    G.add_node(cake.id, label=cake.name, type='cake')

groups = ['chocolate','fruity','creamy','streusel','cake_type']
label = ["Schokoladig", "Fruchtig", "Cremig", "Streuselig", "Hoher Kuchentyp"]
for k, grp in enumerate(groups):
    G.add_node(grp, type='group', label=label[k])

for i, cake in cakes.iterrows():
    for k, grp in enumerate(groups):
        weight = cake[grp]
        if weight > 0:
            G.add_edge(cake['id'], grp, weight=weight)
```

```python

from bokeh.io import show
from bokeh.models import (
    GraphRenderer,
    StaticLayoutProvider,
    Circle,
    MultiLine,
    ColumnDataSource,
    TapTool,
    CustomJS,
    Div,
    LabelSet
)
from bokeh.plotting import figure
from bokeh.layouts import column

# --- 2) Positionen via gewichtetes Spring-Layout ---
pos = nx.spring_layout(
    G,
    weight='weight',
    k=1/np.sqrt(G.number_of_nodes()),
    iterations=100
)

# --- 3) Bokeh DataSources ---
node_indices = list(G.nodes)
node_types   = [G.nodes[n]["type"] for n in node_indices]
node_labels  = [G.nodes[n]["label"] for n in node_indices]
# Farben/Größen pro Typ
node_colors = ["#8ecae6" if t=="cake" else "#ffb703" for t in node_types]
node_sizes  = [0.01 if t=="cake" else 0.02 for t in node_types]

node_source = ColumnDataSource(data=dict(
    index = node_indices,
    type  = node_types,
    label = node_labels,
    color = node_colors,
    size  = node_sizes,
    x     = [pos[n][0] for n in node_indices],
    y     = [pos[n][1] for n in node_indices]
))

starts, ends, weights = zip(*[(u,v,d["weight"]) for u,v,d in G.edges(data=True)])
edge_source = ColumnDataSource(data=dict(
    start   = starts,
    end     = ends,
    weight  = weights,
    visible = [0.5]*len(starts)
))

# --- 4) GraphRenderer ---
graph = GraphRenderer()
graph.node_renderer.data_source = node_source
graph.node_renderer.glyph = Circle(radius="size", fill_color="color", line_color="black")
graph.edge_renderer.data_source = edge_source
graph.edge_renderer.glyph = MultiLine(line_alpha="visible", line_width=2, line_color="black")

graph.layout_provider = StaticLayoutProvider(graph_layout=pos)

# --- 5) Labels für alle Knoten ---
labels = LabelSet(
    x="x", y="y", text="label", source=node_source,
    text_font_size="10pt", x_offset=5, y_offset=5
)

# --- 6) Info-Div für Klick-Ausgabe ---
info = Div(text="Klicke auf einen Kuchen", width=400, height=100)

# --- 7) TapTool + CustomJS Callback ---
callback = CustomJS(args=dict(
    ns=node_source, es=edge_source, info=info
), code="""
    const sel = ns.selected.indices;
if (sel.length===0) {
    info.text = "Klicke auf einen Knoten";
    return;
}
const i = sel[0];
const nid = ns.data['index'][i];
const ntype = ns.data['type'][i];
const nlabel = ns.data['label'][i];

// Arrays für Edge-Quelle/Ziel
const S = es.data['start'], E = es.data['end'];

// Hilfs-Funktion: Filtere Kanten und sammele Nachbarn
function filterAndList(nodeId, nodeType) {
    const others = [];
    for (let j=0; j<S.length; j++) {
        const match = (S[j]===nodeId || E[j]===nodeId);
        es.data['visible'][j] = 0.8 * match + 0.07;
        if (match) {
            const other = (S[j]===nodeId ? E[j] : S[j]);
            const idxOther = ns.data['index'].indexOf(other);
            others.push(ns.data['label'][idxOther]);
        }
    }
    es.change.emit();
    return others;
}

if (ntype === 'cake') {
    // Klick auf Kuchen: zeige Gruppen
    const groups = filterAndList(nid, 'group');
    info.text = `<b>${nlabel}</b><br>Gruppen: ${groups.join(', ')}`;
}
else if (ntype === 'group') {
    // Klick auf Gruppe: zeige Kuchen
    const cakes = filterAndList(nid, 'cake');
    info.text = `<b>${nlabel}</b><br>Kuchen: ${cakes.join(', ')}`;
}
else {
    // fallback, falls andere Typen dazu kommen
    es.data['visible'] = es.data['start'].map(()=>true);
    es.change.emit();
    info.text = "Bitte klicke auf einen Kuchen oder eine Kategorie.";
}
""")
tap = TapTool(callback=callback, renderers=[graph.node_renderer])

# --- 8) Plot zusammenbauen ---
p = figure(title="Kuchen–Attribut-Bipartit-Graph",
           tools=[tap, "pan", "wheel_zoom", "reset"],
           width=1000, height=1000)
p.renderers.extend([graph, labels])

# save the plot to an html file
from bokeh.io import output_file, save
output_file(filename="kuchen_graph.html", title="Kuchenraum")
save(p)
```
